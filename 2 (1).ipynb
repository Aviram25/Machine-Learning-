##### Boston

```python
boston = pd.read_csv('Boston.csv')
x = boston.drop('medv',axis=1)
y = boston['medv']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=25)
```

```python
poly = PolynomialFeatures(degree=2,include_bias=False).set_output(transform='pandas')
x_poly = poly.fit_transform(x_train)
lr = LinearRegression()
lr.fit(x_poly,y_train)
x_poly
```

```python
x1_poly = poly.transform(x_test)
y_pred = lr.predict(x1_poly)
x1_poly
```

```python
r2 = r2_score(y_test, y_pred)
r2
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
sir_ka_ghar = pd.read_csv('Housing.csv')
```

```python
from sklearn.preprocessing import OneHotEncoder
sir_ka_ghar_part2 = sir_ka_ghar[sir_ka_ghar.columns[sir_ka_ghar.dtypes==object]]
sir_ka_ghar_part3 = sir_ka_ghar[sir_ka_ghar.columns[sir_ka_ghar.dtypes!=object]]
```

```python
ohe = OneHotEncoder(drop='first',sparse_output=False).set_output(transform="pandas")
ohe_ghar = ohe.fit_transform(sir_ka_ghar_part2)
ohe.fit_transform(sir_ka_ghar)
```

```python
ghar = pd.concat([sir_ka_ghar_part3,ohe_ghar],axis=1)
x,y = ghar.drop('price',axis=1),ghar['price']
```

```python
def poly_lr(n):
    x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=25)
    poly = PolynomialFeatures(degree=n, include_bias=False).set_output(transform='pandas')
    x_poly = poly.fit_transform(x_train)
    lr = LinearRegression()
    lr.fit(x_poly, y_train)

    x1_poly = poly.transform(x_test)
    y_pred = lr.predict(x1_poly)

    mae = mean_absolute_error(y_test, y_pred)
    print(mae)
```

```python
poly_lr(1)
poly_lr(2)
poly_lr(3)
poly_lr(4)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False).set_output(transform='pandas')
col_trnf = ColumnTransformer([('OHE', ohe, list(sir_ka_ghar.columns[sir_ka_ghar.dtypes==object]))],
                             remainder='passthrough',
                             verbose_feature_names_out=False)
col_trnf = col_trnf.set_output(transform='pandas')
col_trnf.fit_transform(sir_ka_ghar)
```
